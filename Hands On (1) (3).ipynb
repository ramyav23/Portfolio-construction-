```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm

# Training data

X_train = np.array([[1.1,2.1], [-1.1,2.1], [-1.1,-2.1]])
y_train = np.array([-1, -1, 1])


# Create and train the SVM model
model = svm.SVC(kernel='linear', C=1)
model.fit(X_train, y_train)

# Model parameters
w = model.coef_[0]
b = model.intercept_[0]
support_vectors = model.support_vectors_

# Calculate margin width
margin_width = 2 / np.linalg.norm(w)

# Calculate the value of b
b_star = -model.intercept_[0] / w[1]

# Print the model parameters
print(f"Optimal weight vector (w*): {w}")
print(f"Optimal bias (b*): {b_star}")
print(f"Margin width: {margin_width}")

# Print the alpha values
alpha_values = model.dual_coef_[0]
for i, alpha in enumerate(alpha_values):
    print(f"Alpha {i + 1}: {alpha}")

# Predict the class for point D (3, 13)
X_test = np.array([[1.1,1.1]])
prediction_12 = model.predict(X_test)

print(f"Prediction for point D (1,1): {prediction_12}")

# Plot the data points
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.Paired, marker='o', s=100)

# Plot the decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])

# Plot decision boundary and margins
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

# Highlight support vectors
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=300, linewidth=1, facecolors='none', edgecolors='k', label='Support Vectors')

# Plot the hyperplane
plt.plot(xx[0], -(w[0] * xx[0] + b) / w[1], color='cyan', linestyle='-', linewidth=2, label='Hyperplane')

# Plot the margin lines
plt.plot(xx[0], -(w[0] * xx[0] + b - 1) / w[1], 'k--', label='Margin')
plt.plot(xx[0], -(w[0] * xx[0] + b + 1) / w[1], 'k--')

# Highlight test point
plt.scatter(X_test[:, 0], X_test[:, 1], c='red', marker='x', s=200, label='Test Point')

# Set labels and show the plot
plt.title('SVM Decision Boundary with Support Vectors, Hyperplane, and Margin')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree

# Training data with two features

X_train = np.array([[1.1,2.1], [-1.1,2.1], [-1.1,-2.1]])
y_train = np.array([-1, -1, 1])

# Testing data with two features
X_test = np.array([[1.1,1.1]])

# Create a Random Forest classifier with a single decision tree
rf_classifier = RandomForestClassifier(n_estimators=1, max_depth=5, random_state=45)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Access the decision tree
decision_tree = rf_classifier.estimators_[0]

# Define a function to recursively print split features
def print_tree_structure(tree, node, depth):
    indent = "  " * depth
    if tree.tree_.children_left[node] != tree.tree_.children_right[node]:
        print(f"{indent}Decision Node: Split on Feature {tree.tree_.feature[node]} <= {tree.tree_.threshold[node]}")
        print_tree_structure(tree, tree.tree_.children_left[node], depth + 1)
        print_tree_structure(tree, tree.tree_.children_right[node], depth + 1)
    else:
        print(f"{indent}Leaf Node: Class Distribution = {tree.tree_.value[node]}")

# Plot the decision tree without Gini index calculations
plt.figure(figsize=(12, 8))
plot_tree(decision_tree, feature_names=['Feature 1', 'Feature 2'], class_names=['Class -1', 'Class 1'], filled=True, rounded=True, impurity=False, node_ids=True)
plt.title('Decision Tree ')
plt.show()

# Call the function to print split features
print_tree_structure(decision_tree, 0, 0)

# Predict the class of the test point
prediction = rf_classifier.predict(X_test)
print(f"Predicted class for point [1,1]: {prediction[0]}")
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize

# Function to calculate the negative Sharpe ratio (as scipy minimizes)
def negative_sharpe_ratio(weights, returns, risk_free_rate):
    portfolio_return = np.dot(returns.mean(), weights)
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov(), weights)))
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_volatility
    return -sharpe_ratio

# Function to ensure the sum of weights is equal to 1
def weight_constraint(weights):
    return np.sum(weights) - 1.0

# Example data (replace this with your stock returns)
data = {
    'AAPL':APPL_Returns,
    'MSFT': MSFT_Returns,
 
}

# Convert data to DataFrame
returns_df = pd.DataFrame(data)

# Risk-free rate (you may adjust this)
risk_free_rate = 0.01

# Number of assets in the portfolio
num_assets = len(returns_df.columns)


# Initial equal weights for each asset
initial_weights = np.ones(num_assets) / num_assets

# Constraints
constraints = ({'type': 'eq', 'fun': weight_constraint})

# Bounds for weights (0 to 1 for each asset)
bounds = tuple((0, 1) for asset in range(num_assets))

# Optimize the portfolio
result = minimize(negative_sharpe_ratio, initial_weights, args=(returns_df, risk_free_rate),
                  method='SLSQP', bounds=bounds, constraints=constraints)

# Extract the optimized weights
optimized_weights = result.x

# Display the optimized weights
print("Optimal Weights:")
for i in range(num_assets):
    print(f"{returns_df.columns[i]}: {optimized_weights[i]:.4f}")

# Calculate the portfolio return, volatility, and Sharpe ratio
portfolio_return = np.dot(returns_df.mean(), optimized_weights)
portfolio_volatility = np.sqrt(np.dot(optimized_weights.T, np.dot(returns_df.cov(), optimized_weights)))
sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_volatility

print(f"\nPortfolio Return: {portfolio_return:.4f}")
print(f"Portfolio Volatility: {portfolio_volatility:.4f}")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
```
